```python
attribute = 'balance'
target_variables = bank_data.deals.unique()  #This gives all 'Yes' and 'No'
variables = bank_data[attribute].unique()    #This gives different features in that attribute (like 'Sweet')
entropy_attribute = 0
for variable in variables:
    entropy_each_feature = 0
    for target_variable in target_variables:
        num = len(bank_data[attribute][bank_data[attribute]==variable][bank_data.deals ==target_variable]) #numerator
        den = len(bank_data[attribute][bank_data[attribute]==variable])  #denominator
        fraction = num/(den+eps)  #pi
        entropy_each_feature += -fraction*log(fraction+eps) #This calculates entropy for one feature like 'Sweet'
    fraction2 = den/len(bank_data)
    entropy_attribute += -fraction2*entropy_each_feature   #Sums up all the entropy ETaste
```

```python
def find_entropy(bank_data):
    Class = bank_data.keys()[-1]   #To make the code generic, changing target variable class name
    entropy = 0
    values = bank_data[Class].unique()
    for value in values:
        fraction = bank_data[Class].value_counts()[value]/len(bank_data[Class])
        entropy += -fraction*np.log2(fraction)
    return entropy
```

```python
def find_entropy_attribute(bank_data,attribute):
  Class = bank_data.keys()[-1]   #To make the code generic, changing target variable class name
  target_variables = bank_data[Class].unique()  #This gives all 'Yes' and 'No'
  variables = bank_data[attribute].unique()    #This gives different features in that attribute (like 'Hot','Cold' in Temperature)
  entropy2 = 0
  for variable in variables:
      entropy = 0
      for target_variable in target_variables:
          num = len(bank_data[attribute][bank_data[attribute]==variable][bank_data[Class] ==target_variable])
          den = len(bank_data[attribute][bank_data[attribute]==variable])
          fraction = num/(den+eps)
          entropy += -fraction*log(fraction+eps)
      fraction2 = den/len(bank_data)
      entropy2 += -fraction2*entropy
  return abs(entropy2)
```

```python
def find_winner(bank_data):
    Entropy_att = []
    IG = []
    for key in bank_data.keys()[:-1]:
#         Entropy_att.append(find_entropy_attribute(df,key))
        IG.append(find_entropy(bank_data)-find_entropy_attribute(bank_data,key))
    return bank_data.keys()[:-1][np.argmax(IG)]
```

```python
def get_subtable(bank_data, node,value):
  return bank_data[bank_data[node] == value].reset_index(drop=True)
```

```python
def buildTree(bank_data,tree=None): 
    Class = bank_data.keys()[-1]   #To make the code generic, changing target variable class name
    
    #Here we build our decision tree

    #Get attribute with maximum information gain
    node = find_winner(bank_data)
    
    #Get distinct value of that attribute e.g Salary is node and Low,Med and High are values
    attValue = np.unique(bank_data[node])
    
    #Create an empty dictionary to create tree    
    if tree is None:                    
        tree={}
        tree[node] = {}
    
   #We make loop to construct a tree by calling this function recursively. 
    #In this we check if the subset is pure and stops if it is pure. 

    for value in attValue:
        
        subtable = get_subtable(bank_data,node,value)
        clValue,counts = np.unique(subtable['deals'],return_counts=True)                        
        
        if len(counts)==1:#Checking purity of subset
            tree[node][value] = clValue[0]                                                    
        else:        
            tree[node][value] = buildTree(subtable) #Calling the function recursively 
                   
    return tree
  
  
```

```python
tree = buildTree(bank_data)
```

```python
def predict(inst,tree):
    #This function is used to predict for any input variable 
    
    #Recursively we go through the tree that we built earlier

    for nodes in tree.keys():        
        
        value = inst[nodes]
        tree = tree[nodes][value]
        prediction = 0
            
        if type(tree) is dict:
            prediction = predict(inst, tree)
        else:
            prediction = tree
            break;                            
        
    return prediction
```
